### Subtask 2.1: Using 100 Covid-19 and 100 Normal CXRs images

```python
# Step 1: Creating directories
import os, shutil
original_data_dir = '/content/gdrive/MyDrive/cgan_Assessment'

base_dir = '/content/base_dir'
os.mkdir(base_dir)

train_dir = os.path.join(base_dir, 'train')
val_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')
os.mkdir(train_dir)
os.mkdir(val_dir)
os.mkdir(test_dir)

train_covid_dir = os.path.join(train_dir, 'covid')
train_normal_dir = os.path.join(train_dir, 'normal')
os.mkdir(train_covid_dir)
os.mkdir(train_normal_dir)

val_covid_dir = os.path.join(val_dir, 'covid')
val_normal_dir = os.path.join(val_dir, 'normal')
os.mkdir(val_covid_dir)
os.mkdir(val_normal_dir)

test_covid_dir = os.path.join(test_dir, 'covid')
test_normal_dir = os.path.join(test_dir, 'normal')
os.mkdir(test_covid_dir)
os.mkdir(test_normal_dir)
```

```python
# Step 2: Copy images from Google Drive
import glob
import os
import shutil
from tqdm import tqdm

original_covid_dir = os.path.join(original_data_dir, 'Covid-19')
original_normal_dir = os.path.join(original_data_dir, 'Normal')

covidImgs = os.listdir(original_covid_dir)
normalImgs = os.listdir(original_normal_dir)

# copy 60 covid images and 60 normal images to train dirctory (total 120)
fnames = [a for a in covidImgs[:60]]
print('copy files to train_covid_dir')
for fname in tqdm(fnames):
    src = os.path.join(original_covid_dir, fname)
    dst = os.path.join(train_covid_dir, fname)
    shutil.copyfile(src, dst)

fnames = [a for a in normalImgs[:60]]
print('copy files to train_normal_dir')
for fname in tqdm(fnames):
    src = os.path.join(original_normal_dir, fname)
    dst = os.path.join(train_normal_dir, fname)
    shutil.copyfile(src, dst)

# copy 20 covid images and 20 normal images to validation directory (total 40)
fnames = [a for a in covidImgs[60:80]]
print('copy files to train_normal_dir')
for fname in tqdm(fnames):
    src = os.path.join(original_covid_dir, fname)
    dst = os.path.join(val_covid_dir, fname)
    shutil.copyfile(src, dst)
fnames = [a for a in normalImgs[60:80]]
print('copy files to train_normal_dir')
for fname in tqdm(fnames):
    src = os.path.join(original_normal_dir, fname)
    dst = os.path.join(val_normal_dir, fname)
    shutil.copyfile(src, dst)

# copy 20 covid images and 20 normal images to test directory (total 40)
fnames = [a for a in covidImgs[80:]]
print('copy files to train_normal_dir')
for fname in tqdm(fnames):
    src = os.path.join(original_covid_dir, fname)
    dst = os.path.join(test_covid_dir, fname)
    shutil.copyfile(src, dst)
fnames = [a for a in normalImgs[80:]]
print('copy files to train_normal_dir')
for fname in tqdm(fnames):
    src = os.path.join(original_normal_dir, fname)
    dst = os.path.join(test_normal_dir, fname)
    shutil.copyfile(src, dst)
```

```python
# Step 3: Import ResNet50
from keras.applications.resnet import ResNet50

conv_base = ResNet50(weights='imagenet',
                     include_top=False,
                     input_shape=(150,150,3))
conv_base.summary()
```

```python
# Step 4: Add Dense Layers
from keras import models
from keras import layers

model = models.Sequential()
model.add(conv_base)
model.add(layers.Flatten())
model.add(layers.Dense(1024, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.summary()
# freeze the whole resnet50 model
conv_base.trainable = False
```

```python
# Step 5 and 6: Data Argumentation and training the model
from keras.preprocessing.image import ImageDataGenerator
from keras import optimizers

train_datagen = ImageDataGenerator(
    rescale = 1./255,
    rotation_range = 45,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    fill_mode = 'nearest'
)

# We don't have to process the test dataset
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size = (150, 150),
    batch_size = 20, 
    class_mode = 'binary'
)

validation_generator = test_datagen.flow_from_directory(
    val_dir,
    target_size = (150, 150),
    batch_size = 20,    
    class_mode = 'binary'
)

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=2e-5),
              metrics=['acc'])

history = model.fit_generator(
    train_generator,
    steps_per_epoch=6, # amendable
    epochs = 250,
    validation_data = validation_generator,
    validation_steps=2 # amendable
)
```

```python
# Step 7: Draw the image
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs=range(1, len(acc)+1)

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation_acc')
plt.title('Training and valication accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'ro', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Valication loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.show()
```

```python
def smooth_curve(points, factor=0.8):
    smoothed_points=[]
    for point in points:
        if smoothed_points:
            previous = smoothed_points[-1]
            smoothed_points.append(previous * factor + point*(1-factor))
        else:
            smoothed_points.append(point)
    return smoothed_points
```

```python
plt.plot(epochs, smooth_curve(acc), 'bo', label='Smoothed_training_acc')
plt.plot(epochs, smooth_curve(val_acc), 'b', label='Smoothed_validation_acc')
plt.title("Training and validation accuracy")
plt.legend()

plt.figure()

plt.plot(epochs, smooth_curve(loss), 'ro', label='Smoothed_training_loss')
plt.plot(epochs, smooth_curve(val_loss), 'r', label='Smoothed_validation_loss')
plt.title("Training and validation loss")
plt.legend()

plt.show()
```

```python
# train the model over again
history = model.fit_generator(
    train_generator,
    steps_per_epoch=6, # amendable
    epochs = 235,
    validation_data = validation_generator,
    validation_steps=2 # amendable
)
```

```python
# evaluate the data with test data set
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150,150),
    batch_size=20,
    class_mode='binary'
)
test_loss, test_acc = model.evaluate(test_generator, steps=50)
```

```python
print('Test loss: ', test_loss)
print('Test acc: ', test_acc)
```
